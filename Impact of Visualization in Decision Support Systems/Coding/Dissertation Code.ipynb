```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for seaborn
sns.set_style("darkgrid")

# Create a bar plot to visualize the average Price by County
plt.figure(figsize=(16, 6))  # Increase the figure width to accommodate more space for labels
ax = sns.barplot(data=df, x='County', y='Price', ci=None)
plt.title("Average Price by County")
plt.xlabel("County")
plt.ylabel("Average Price")
plt.xticks(rotation=90, ha='right', fontsize=8)  # Adjust rotation, alignment, and font size
plt.tight_layout()  # Adjust the spacing between subplots for better readability

# Adjust the y-axis label to improve visibility
ax.yaxis.set_label_coords(-0.07, 0.5)

plt.show()
```
